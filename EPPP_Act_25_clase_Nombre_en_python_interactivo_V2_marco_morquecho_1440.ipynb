Crear la clase Calificacion, con 3 atributos (alumno, materia, nota), 2 funciones leerdatos(), mostrardatos() y 2 objetos (calif1, calif2).

```python
# Act25 Clase Calificacion Marco 1440
# Identificación: Marco Morquecho | ID: 1440 | NL: 39
# Descripción: Clase que gestiona calificaciones de alumnos de forma interactiva.

class Calificacion:
    # Atributos: alumno, materia, nota (se inicializan en leerdatos)

    # Función 1: Leer datos desde el teclado
    def leerdatos(self):
        print("\n--- Captura de Nueva Calificación ---")
        self.alumno = input("Nombre del alumno: ")
        self.materia = input("Nombre de la materia: ")
        self.nota = input("Calificación obtenida: ")

    # Función 2: Mostrar los datos capturados
    def mostrardatos(self):
        print("\n" + "·"*30)
        print(f"ALUMNO:  {self.alumno}")
        print(f"MATERIA: {self.materia}")
        print(f"NOTA:    {self.nota}")
        print("·"*30)

# --- Creación de 2 Objetos (Instancias) ---
print("SISTEMA DE CONTROL ESCOLAR V2")

calif1 = Calificacion()
calif1.leerdatos()

calif2 = Calificacion()
calif2.leerdatos()

# --- Resultados finales ---
print("\n=== REPORTE DE CALIFICACIONES ===")
calif1.mostrardatos()
calif2.mostrardatos()

print("\nPrograma realizado por Marco Morquecho NL=39 1440")
```
